```python
# Importing necessary libraries
import pandas as pd  # For data manipulation
import numpy as np  # For numerical operations
import matplotlib.pyplot as plt  # For visualization
import seaborn as sns  # For better-looking plots

# Setting styles for plots
sns.set(style="darkgrid")

# Ignore warnings (optional)
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
df = pd.read_csv("googleplaystore (1).csv")

# Display the first few rows to understand the data
df.head()
```

```python
# Get basic information about the dataset
df.info()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Summary of numerical columns
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df.fillna(value="Unknown", inplace=True)  # Example for categorical columns
```

```python
df.dtypes
```

```python
df['Reviews'] = df['Reviews'].astype(int)  # Convert to integer
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)  # Clean and convert
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
df.head()
```

```python
df.describe()
```

```python
df['Category'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(y=df['Category'], order=df['Category'].value_counts().index)
plt.title('Number of Apps per Category')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x=df['Reviews'])
plt.title("Boxplot of Reviews")
plt.show()
```

```python
top_categories = df['Category'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_categories.index, y=top_categories.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Top 10 App Categories by Count")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.show()
```

```python
top_installs = df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_installs.index, y=top_installs.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Categories by Total Installs")
plt.xlabel("Category")
plt.ylabel("Total Installs")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Rating'], bins=20, kde=True, color='blue')
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Type'], y=df['Rating'], palette="pastel")
plt.title("Rating Distribution: Free vs Paid Apps")
plt.show()
```

```python
plt.figure(figsize=(15,10))

# Subplot 1: Top Categories by Number of Apps
plt.subplot(2,2,1)
sns.barplot(x=top_categories.index, y=top_categories.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Top 10 Categories by Number of Apps")

# Subplot 2: Top Categories by Installs
plt.subplot(2,2,2)
sns.barplot(x=top_installs.index, y=top_installs.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Categories by Total Installs")

# Subplot 3: Ratings Distribution
plt.subplot(2,2,3)
sns.histplot(df['Rating'], bins=20, kde=True, color='blue')
plt.title("Distribution of App Ratings")

# Subplot 4: Free vs Paid Ratings
plt.subplot(2,2,4)
sns.boxplot(x=df['Type'], y=df['Rating'], palette="pastel")
plt.title("Free vs Paid Apps - Ratings Comparison")

plt.tight_layout()
plt.show()
```

# Google Play Store Apps Analysis  

## 📌 Objective  
This project analyzes Google Play Store apps to extract insights on:  
- **Category distribution**: Most common app categories.  
- **Install trends**: Categories with the highest number of downloads.  
- **User ratings**: Understanding rating distribution and trends.  
- **Pricing impact**: Comparing free vs. paid apps in terms of ratings and popularity.  

## 📊 Key Findings  
1️⃣ **Most Common App Categories**:  
   - The most frequent app categories are **Family (1939 apps), Games (1121 apps), and Tools (841 apps)**.  

2️⃣ **Highest Installed Categories**:  
   - **Games (31.5 billion installs), Communication (24.1 billion installs), and Social (12.5 billion installs)** lead in total downloads.  

3️⃣ **Ratings Analysis**:  
   - The **average rating is 4.19**, indicating overall user satisfaction.  
   - Ratings show a **slight right-skew**, meaning most apps are rated positively.  

4️⃣ **Free vs. Paid Apps**:  
   - **Paid apps tend to have higher ratings** than free apps, suggesting better quality or user experience.  

## 🔍 Recommendations  
✅ **Developers should focus on Family, Games, and Tools categories** for new app launches.  
✅ **To maximize downloads, creating apps in Games, Communication, and Social categories is ideal**.  
✅ **Paid apps are generally worth investing in** as they receive higher ratings, indicating user satisfaction.
